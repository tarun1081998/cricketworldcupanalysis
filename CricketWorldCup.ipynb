```python
# TO GET THE LINE NUMBER WHERE OUR REQUIRED DATA IS STARTING
# USING REGULAR EXPRESSION TO GET LINE NUMBER FROM OUR ERROR USING EXCEPTION HANDLING
def get_line(a):
    try:
        pd.read_csv(a)
    except Exception as e:
        a=str(e)
        x=re.findall(r'[l]\w+ \d+',a)
        return(int((re.findall(r'\d+',x[0]))[0]))
```

```python
#TO CREATE THE DATAFRAME
#FIRST WE GET OUR FILE AS THERE CAN BE ANY OTHER FILES ARE PRESENT IN OUR DATASET WE USE REGULAR EXPRESSION , GET THE FILE 
#STARING FROM A NUMBER AS OUR REQUIRED FILE IS STARTING WITH A NUMBER
count=0
df_wc=pd.DataFrame()
for i in x:
    if re.match(r'^\d',i):
        line=0
        line=get_line(i)
        a=pd.read_csv(i,nrows=line-2)
        if 'ICC World Cup' not in a['1.3.0'].values:
            continue
        else:
            df=pd.read_csv(i,skiprows=line-1,index_col=False,header=None)
            df_wc=df_wc.append(df)
            count+=1
```

```python
print('number of ICC World Cup Matches are :',count)
```

```python
def strike_rate(name):
    run=sum((df_wc[df_wc[4].values==name])[7])
    balls=len(df_wc[df_wc[4].values==name])
    return (run/balls)*100
```

```python
strike_rate('CH Gayle')
```

```python
def average(name):
    run=sum((df_wc[df_wc[4].values==name])[7])
    d=df_wc[df_wc[4].values==name]
    d=d.astype(str)
    out=len(d[d[9]!='nan'])
    return run/out
```

```python
average('CH Gayle')
```
